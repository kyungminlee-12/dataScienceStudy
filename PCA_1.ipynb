### 붗꽃 데이터의 차원축소
~~~
붓꽃 데이터 중에서 10송이 데이터 증에서 꽃받침 길이와 꽃받침 폭 데이터를 그래프로 보이면 다음과 같다.
첫 번째 그래프는 가로축을 특성의 종류, 세로 축을 특성의 값으로 나타낸 것이다.
이 그래프에서 꽃받침 길이가 크면 꽃받침 폭도 같이 커진다는 규칙을 알 수 있다.
~~~

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
N=10   #앞의 10송이만 선택
X=iris.data[:N, :2]        #10개의 꽃의 특징 2가지 갖고옴

plt.plot(X.T, 'o:')        #X.T는 X의 전치행렬 (행과 열 바꿈), 그래프를 그릴 데아터(숫자)를 input, 뒤에는 스타일을 지정한 것 같음
plt.xticks(range(4), ['Calyx length', 'Calyx width'])     # x축의 내용: ['꽃받침 길이', '꽃받침 폭']
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title('characteristic of iris size')     #붓꽃 크기의 특성 
plt.legend(['sample {}'.format(i+1) for i in range(N)])    #항목 표시
plt.show()
```

### 가로축이 꽃받침 길이, 세로축이 꽃받침 폭으로 하는 스캐터 플롯
~~~
위의 그래프와 유사하게 꽃받침 길이가 크면 꽃받침 폭도 같이 커진다는 규칙을 알 수 있다
~~~

```python
import seaborn as sns
import pandas as pd
```

```python
plt.figure(figsize=(8, 8))    #틀 사이즈 지정
ax=sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color='.2', marker='s')      #그래프 형식 지정(scatterplot)
#s: marker size
#scatterplot(산점도): 직교 좌표계를 이용해 두 개 변수 간의 관계를 나타낸다
#0, 1의 의미는 정확하게는 모르겠...

for i in range(N):
    ax.text(X[i, 0]-0.05, X[i, 1]+0.03, 'sample {}'.format(i+1))

plt.xlabel('Calyx length')        # '꽃받침 길이'
plt.ylabel('Calyx width')         # '꽃받침 폭'
plt.title('characteristic of iris size (2 dimension)')
plt.axis('equal')
plt.show()
```
