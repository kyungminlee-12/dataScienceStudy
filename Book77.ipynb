```python
import matplotlib.pyplot as plt
from scipy import misc   
```

```python
img_rgb=misc.face()
img_rgb.shape
```

```python
plt.subplot(221)
plt.imshow(img_rgb[:,:,0], cmap=plt.cm.gray)
plt.axis('off')                               #x, y축에 좌표 겂둘이 들어가는 것을 없앰
plt.title('red channel')
```

```python
plt.subplot(221)   #세로 2개 가로 2개 4개의 틀 생성 후 1~4 차례대로 특정 틀 불러옴
plt.imshow(img_rgb, cmap=plt.cm.gray)    #img_rgb에 index 값을 지정하면 특정 색의 pixel값을 불러올 수 있다
plt.axis('off')                          #x, y축에 좌표 겂둘이 들어가는 것을 없앰
plt.title('RGB channel')

plt.subplot(222)
plt.imshow(img_rgb[:,:,0], cmap=plt.cm.gray)
plt.axis('off')                               #x, y축에 좌표 겂둘이 들어가는 것을 없앰
plt.title('Red channel')

plt.subplot(223)
plt.imshow(img_rgb[:,:,1], cmap=plt.cm.gray)
plt.axis('off')                               #x, y축에 좌표 겂둘이 들어가는 것을 없앰
plt.title('Green channel')

plt.subplot(224)
plt.imshow(img_rgb[:,:,2], cmap=plt.cm.gray)
plt.axis('off')                               #x, y축에 좌표 겂둘이 들어가는 것을 없앰
plt.title('Blue channel')
```
