```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=np.array([1, 2])
x2=np.array([2, 1])
x3=0.5*x1+x2
```

```python
#색 설정
#annotate의 벡터를 그리는 함수의 arrowprops라는 keyword에 사용된다.
#꼭 dict 형태로 사용된다. 
gray={'facecolor': 'gray'}         #facecolor라는 것을 gray로 설정하는 명령어
black={'facecolor': 'black'}
red={'facecolor': 'red'}
green={'facecolor': 'green'}
blue={'facecolor': 'blue'}
```

# 벡터의 선형 조합
### data science school pg 146

```python
plt.annotate('', xy=0.5*x1, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x2, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x3, xytext=(0, 0), arrowprops=black)

plt.plot(0, 0, 'kP', ms=10)                 #'kP'는 X 표시를 의미한다.
plt.plot(x1[0], x1[1], 'ro', ms=10)         #'ro'는 빨간색 점을 의미한다, ms는 원의 크기를 조절한다.
plt.plot(x2[0], x2[1], 'ro', ms=10)
plt.plot(x3[0], x3[1], 'ro', ms=10)

plt.plot([x1[0], 0], [x1[1], 0], 'k--')         #'k--'는 검은색 끊어진 실선을 의미한다.
plt.text(0.6, 2.0, '$x_1$')                     
plt.text(-0.5, 0.5, '$0.5x_1$')
plt.text(1.15, 0.25, '$x_2$')
plt.text(2.5, 1.6, '$0.5x_1+x_2$')

plt.xticks(np.arange(-2, 5))              #축에 그려지는 높이 표시 지점들을 지정하는 코드
plt.yticks(np.arange(-1, 4))

plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

## 벡터의 차

```python
a=np.array([1, 2])
b=np.array([2, 1])
c=a-b
```

```python
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)     #annotate는 벡터를 그리는 코드
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)     #xy=는 벡터의 방향을 결정한다.
plt.annotate('', xy=a, xytext=b, arrowprops=black)         #xytext는 벡터의 시작점을 의미
                                 #arrowprops는 dict 형태로 {'facecolor':'원하는 색'}의 방식으로 지정한다.

plt.plot(0, 0, 'kP', ms=10)                               #'kP'는 X 표시를 의미한다.
plt.plot(a[0], a[1], 'ro', ms=10)                         #'ro'는 빨간색 점을 의미한다.
plt.plot(b[0], b[1], 'ro', ms=10)

plt.text(0.35, 1.15, '$a$')
plt.text(1.15, 0.25, '$b$')
plt.text(1.55, 1.65, '$a-b$')

plt.xticks(np.arange(-2, 5))          #xticks는 좌표를 지정한다.
plt.yticks(np.arange(-1, 4))

plt.xlim(-0.8, 2.8)
plt.ylim(-0.8, 2.8)
plt.show()
```

# Word2Vec

~~~
서울에서 한국을 뺀 벡터를 사용해서 도쿄를 수도로 하는 나라를 알아내는데 사용한다.
일본=도쿄+(한국-서울)
~~~

```python
a=np.array([2, 2])
b=np.array([3, 4])
c=np.array([4, 1])
d=a+(c-a)
e=b+(c-a)
```

```python
plt.annotate('', xy=b, xytext=a, arrowprops=black)        #annotate는 백터를 그리는 함수
plt.annotate('', xy=e, xytext=d, arrowprops=black)        #xy는 벡터의 방향을 의미하는 함수
plt.annotate('', xy=c, xytext=[0, 0], arrowprops=gray)    #xytext는 벡터의 시작점을 의미한다.

plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)

plt.text(1.6, 1.5, 'Seoul')
plt.text(2.5, 4.3, 'Korea')
plt.text(3.5, 0.5, 'Tokyo')
plt.text(4.9, 3.2, 'Japan')

plt.xticks(np.arange(-2, 7))
plt.yticks(np.arange(-1, 6))

plt.xlim(-1.4, 6.4)
plt.ylim(-0.6, 5.8)
plt.show()
```
