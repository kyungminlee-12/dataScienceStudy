다음 명령으로 붓꽃의 꽃잎 길이(petal length) 데이터를 구하고 히스토그램을 그려 분포의 모양을 살핀다

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris=load_iris()   #2차원 array의 data
df=pd.DataFrame(iris.data, columns=iris.feature_names)
sy=pd.Series(iris.target, dtype='category')
sy=sy.cat.rename_categories(iris.target_names)
df['species']=sy
```

```python
print(df)
```

```python
print(sy)
```

```python
#다봉분포인 꽃잎 길이
df['petal length (cm)'].hist()       #key word를 의미하는 것 ('petal length (cm)')라는 항목을 histogram으로 표현하기
plt.title('Length of flower leaf')   #title 지정
plt.show()                           #시각화
```

```python
#단봉 분포
df[df.species=='setosa']['petal length (cm)'].hist()
plt.title('Length of the flower leaf')   
plt.show()
```

# 대칭 분포

```python
np.random.seed(1)
x=np.random.normal(size=1000)
x=np.hstack([x, 5*np.ones(50)])
bins=np.linspace(-6, 6, 12*4+1)
ns, _=np.histogram(x, bins=bins)

sample_mean=np.mean(x)
sample_median=np.median(x)
mode_index=np.argmax(ns)
sample_mode=0.5*(bins[mode_index]+bins[mode_index+1])

sns.distplot(x, bins=bins)
plt.axvline(sample_mean, c='k', ls=':', label='mean')
plt.axvline(sample_median, c='k', ls='--', label='median')
plt.axvline(sample_mode, c='k', ls='-', label='mode')
plt.title('difference between mean, median, and mode')
plt.xlabel('x')
plt.legend()
plt.show()
```
