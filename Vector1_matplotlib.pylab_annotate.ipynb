```python
import numpy as np
import matplotlib.pylab as plt
```

# 평행이동

```python
plt.rc('font', size=18)      #그림의 폰트의 크기를 18으로 지정
```

```python
#색 설정
gray={'facecolor': 'gray'}         #facecolor라는 것을 gray로 설정하는 명령어
black={'facecolor': 'black'}
red={'facecolor': 'red'}
green={'facecolor': 'green'}
blue={'facecolor': 'blue'}
```

```python
a=np.array([1, 2])
plt.plot(0, 0, 'kP', ms=20)         #ms: markersize로 마커으 크기를 지정(X 표시의 크기 조정), 'kP'는 X표시를 생성
plt.plot(a[0], a[1], 'ro', ms=20)   #'ro'는 빨간색 원을 생성하는 code, ms=20은 원의 사이즈 조절

plt.annotate('', xy=[-0.6, 1.6], xytext=(0.2, 0.7), arrowprops=gray)   # xy=[-0.6, 1.6]는 벡터의 방향을 설정한다
                                                                       #x의 음의 방향으로 0.6, y의 양의 방행으로 1.6 이동하는 모양의 벡터
                                                                       #xytext=(0.2, 0.7)는 화살표의 시작점을 의마한다.
                                                                       #''는 무슨 의미인지는 모르겠으나 없으면 오류 남

plt.annotate('', xy=a, xytext=(0, 0), arrowprops=black)               #빨간 점과 관련 있는 아래 있는 화살표
plt.annotate('', xy=a+[-1, 1], xytext=(-1, 1), arrowprops=black)      #위에 있는 화살표
plt.annotate('', xy=a+[-1, 1], xytext=(-1, 1), arrowprops=black)      #이 아이는 없어도 아무 문제가 없음!! (책에 오탄가?)

plt.text(0.35, 1.15, '$a$')           # (0.35, 1.15) 위치에 a라고 씀
plt.text(1.15, 2.25, '$(1, 2)$')      # (1.15, 2.25) 위치에 (1, 2)라고 작성
plt.text(-0.7, 2.1, '$a$')            #$는 없어도 결과가 동일함!!!(존재 이유 아직 모름)
plt.text(-0.9, 0.6, 'parallel\n translation')

plt.xticks(np.arange(-2, 4))    # x 좌표를 -2~3까지 표시한다
plt.yticks(np.arange(-1, 4))    # y 좌표를 -1~3까지 표시한다(단, -1이 실제로 표시 안된 이유는 ylim의 제한 때문)

#틀의 시작과 끝 범위를 정한다 (틀이 해당 크기 만큼 그려진다)
plt.xlim(-2.4, 3.4)    
plt.ylim(-0.8, 3.4)
plt.show()         #최종 시각화
```

틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자식을 넣을 수도 있다.

**plt.annotate**
~~~
텍스트를 화살표와 함께 넘겨주는 형시이다
arrowprops: argument를 함께 넘겨서 화살표를 함께 만들어줄 수도 있다
~~~

# 벡터의 길이
~~~
벡터의 길이는 놈으로 계산한다
numpy linalg 서브패키지의 norm() 명령으로 벡터의 길이 계산
~~~

```python
a=np.array([1, 2])
np.linalg.norm(a)      #벡터의 길이를 계산하는 코드: 결과는 루트 5와 동일한 값임
```

# 스칼라와 벡터의 곱

```python
a=np.array([1, 2])
b=2*a
c=-a

plt.annotate('', xy=b, xytext=(0, 0), arrowprops=red)   #xy=b 벡터의 방향 설정, xytext=(0, 0)는 시작점 설정, arrowprops=red
plt.text(0.8, 3.1, '$2a$')                               #text함수는 위치 지정 후에 쓸 내용을 str형태로 가입한다
plt.text(2.2, 3.8, '$(2, 4)$')                          #$를 쓰는 이유를 잘은 모르지만 특수 문자를 쓸 때 앞뒤에 자주 쓴다.

plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.text(0.1, 1.3, '$a$')
plt.text(1.1, 1.4, '$(1, 2)$')

plt.plot(c[0], c[1], 'ro', ms=10)                        #'ro'는 빨간색 원을 그린다.

plt.annotate('', xy=c, xytext=(0, 0), arrowprops=blue)
plt.text(-1.3, -0.8, '$-a$')
plt.text(-3, -2.5, '$(-1, -2)$')

plt.plot(0, 0, 'kP', ms=20)

plt.xticks(np.arange(-5, 6))
plt.yticks(np.arange(-5, 6))

plt.xlim(-4.4, 5.4)
plt.ylim(-3.2, 5.2)
plt.show()
```

# 단위벡터

```python
a=np.array([1, 0])
b=np.array([0, 1])
c=np.array([1/np.sqrt(2), 1/np.sqrt(2)])
np.linalg.norm(a), np.linalg.norm(b), np.linalg.norm(c)   #벡터의 길이 구하는 코드
```

# 벡터의 합

### 방법 1)

```python
a=np.array([1,2])
b=np.array([2, 1])
c=a+b

plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)   #xy는 벡터의 방향 설정, xytext=(0, 0)가 벡터의 시작점
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)

plt.plot(0, 0, 'kP', ms=10)                          #'kP'는 X표시를 나타낸다
plt.plot(a[0], a[1], 'ro', ms=10)                    #'ro'는 빨간색 원을 의미한다
plt.plot(b[0], b[1], 'ro', ms=10)                    #ms는 특정 기호나 도형의 크기 조절 시 사용한다.
plt.plot(c[0], c[1], 'ro', ms=10)

plt.plot([a[0], c[0]], [a[1], c[1]],'k--' )   # 'k--': 검은색 점선
plt.plot([b[0], c[0]], [b[1], c[1]],'k--' )

plt.text(0.35, 1.15, '$a$')   
plt.text(1.15, 1.45, '$b$')
plt.text(1.25, 1.45, '$c$')

plt.yticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))

plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)

plt.show()
```

### 방법 2)

```python
a=np.array([1,2])
b=np.array([2, 1])
c=a+b

plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)   #xy는 벡터의 방향 설정, xytext=(0, 0)가 벡터의 시작점
plt.annotate('', xy=c, xytext=a, arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)

plt.plot(0, 0, 'kP', ms=10)                      #'kP'는 X표시를 의미한다
plt.plot(a[0], a[1], 'ro', ms=10)                #'ro'는 빨간색 원을 의미, ms는 원의 크기를 지정할 때 사용 
plt.plot(c[0], c[1], 'ro', ms=10)

plt.text(0.35, 1.15, '$a$')   
plt.text(1.45, 2.45, '$b$')
plt.text(1.25, 1.45, '$c$')

plt.yticks(np.arange(-2, 5))        #yticks는 좌표 축의 단위를 설정할 때 지정
plt.yticks(np.arange(-1, 4))

plt.xlim(-1.4, 4.4)                #xlim은 축의 단위를 지정할 때 사용한다.
plt.ylim(-0.6, 3.8)

plt.show()
```

## annotate을 사용하면 그램 내에 화살표를 포함한 부가 설명 가능

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, S, label="sine")
plt.scatter([0], [0], color="r", linewidth=10)
plt.annotate(r'$(0,0)$', xy=(0, 0), xycoords='data', xytext=(-50, 50),    #xy는 벡터의 방향, tytext는 벡터의 시작점
             textcoords='offset points', fontsize=16, 
             arrowprops=dict(arrowstyle="->", linewidth=3, color="g"))    #arrowprops는 항상 dict형태로 지정하는 것을 알 수 있다
plt.show()
```
