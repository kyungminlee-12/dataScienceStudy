## plt.text 설정하기
~~~
fontdict 키워드를 이용하면 font의 종류, 크기, 색상, 투명도, weight 등의 텍스트 스타일을 설정할 수 있다.
rotation 키워드를 이용해서 텍스트를 회전할 수 있습니다.
~~~

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)      #표준 정규분포 숫자 random으로 10000개
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family': 'serif',                #글씨체 지정 
      'color':  'darkred',                 #글자 색 지정
      'weight': 'normal',                  #글자 굵기 ('normal', 'bold')
      'size': 16}                          #글자의 크기 지정

font2 = {'family': 'Times New Roman',      #글자체
      'color':  'blue',                    #글자 색 지정
      'weight': 'bold',                    #글자 굵기
      'size': 12,                          #글자 크기 지정
      'alpha': 0.7}                        #투명도 지정

font3 = {'family': 'Arial',                #글자체
      'color':  'forestgreen',             #글자 색 지정
      'style': 'italic',
      'size': 14}

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(1.0, 0.35, 'font_1: np.random.randn()', fontdict=font1)

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.20, 'font_2: np.random.standard_normal()', fontdict=font2)

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'font_3: np.random.rand()', fontdict=font3)

plt.show()
```

### text 회전하기

```python
#bins 개수는 막대 개수 조절, alpha는 투명도 조절 
plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')      #histtype은 bar의 디자인을 지정
                                                                      #histtype='step'은 가운데가 색칠되있지 않고 비워짐
plt.text(-3.0, 0.15, 'np.random.randn()', fontdict=font1, rotation=85)

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')   #histtype='stepfilled'은 가운데가 색칠이 되어있음
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60)

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')    #density는 정규화하기 위한 코드이다.
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3)

plt.show()
```

**참고 hist() 사용법**
~~~
hist(x, bins=None, range=None, density=None, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, normed=None, *, data=None, **kwargs)[source]
~~~
**histtype : {'bar', 'barstacked', 'step', 'stepfilled'}, optional**
~~~
histogram 그림의 종류 지정
    'bar'은 bar 옆에 또하나의 bar가 등장하는 가장 기본적인 디자인 .
    'barstacked' is a bar-type histogram (여러개의 data 등장 경우 하나 bar위에 또 다른 bar 올라감.
    'step' generates a lineplot (가운데가 비어있고 테두리 만 그려진 디자인).
    'stepfilled' generates a lineplot (가운데가 다 색칠된 디자인-이 디자인은 bar와의 차이점을 아직 찾지 못함).
    Default is 'bar'
~~~    
**density : bool, optional**
~~~
만약 True일 경우 반환하는 투플에서 1st요소들의 값들을 확룰밀도를 형성시키기 위해 정구화 시켜(=히스토그램 값들의 면적, 적분 값은 1이 된다)
이것은 요소의 수로 나누어 전규화되는 것이 아니라 막대(bin)의 너비를 이용하여 정규화되므로 이에 유의하도록 한다.
만약에 sracked 역시 True일 경우, 히스토그램의 합은 1로 정규화된.)

이 크워드 인자와 normed의 가본 값은 None이며, 만약 둘 중 하나라도 입력이 된다면, 그 값을 사용한다. 만약 둘 다 입력이 되지 않는다면, 그 때는 False로 여겨진다.

normed와 density 둘다 동시에 입력할 경우에는 error가 발생하므로 주의하기

### text 박스 설정하기
~~~
bbox 키워드를 사용해서 텍스트 상자의 스타일을 설정할 수 있습니다.
예제에서는 ‘boxstyle’, ‘ec’, ‘fc’, ‘linestyle’, ‘linewidth’ 등의 스타일을 사용했습니다.
  ‘ec’는 ‘edgecolor’, ‘fc’는 ‘facecolor’와 같습니다. 각각 텍스트 상자의 테두리와 면의 색을 지정하는 속성입니다.
~~~

```python
box1 = {'boxstyle': 'round',
        'ec': (1.0, 0.5, 0.5),
        'fc': (1.0, 0.8, 0.8)}

box2 = {'boxstyle': 'square',
        'ec': (0.5, 0.5, 1.0),
        'fc': (0.8, 0.8, 1.0),
        'linestyle': '--'}

box3 = {'boxstyle': 'square',
        'ec': (0.3, 1.0, 0.5),
        'fc': (0.8, 1.0, 0.5),
        'linestyle': '-.',
        'linewidth': 2}

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(-3.0, 0.15, 'np.random.randn()', fontdict=font1, rotation=85, bbox=box1)
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60, bbox=box2)
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3, bbox=box3)
plt.show()
```
