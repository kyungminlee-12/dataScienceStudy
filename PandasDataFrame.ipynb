**Pandas에 대표적인 indexing 방법들**
~~~
.loc[label]: 레이블을 기반으로 인덱싱을 한다
.iloc[position]: 위치 정수를 기반으로 인덱싱을 한다
loc과 iloc는 row를 선택한다. 이는 column을 선택하는 []와 다르다

.at[label]: 레이블을 기반으로 인덱싱을 한다. (dataframe의 series에 상관없이 하나의 스칼라 값에 접근)
.iat[position]: 위치 정수를 기반으로 인덱싱 (dataframe에 series에 상관없이 하나의 스칼라 값에 접근)
~~~

```python
train_data.loc[1000]
```

```python
train_data.loc[[105,150,1105]]
```

```python
train_data.loc[100]
```

```python
#iloc과 loc의 차의점
train_data.iloc[0]   #바뀐 index가 아닌 기존의 0부터 시작하는 index를 기준한다
```

# dataframe에 새 컬럼 추가 & 삭제

**새 column 추가하기**

[ ]를 사용하거나 insert 함수를 사용하여 원하는 위치에 추가하기

```python
train_data['new_fare_amount']=train_data['fare_amount']+2
train_data.head()
```

```python
#(dataset).insert(insert_index, 'inserted column name', insert values)
train_data.insert(4, 'average_longitude', ((train_data['pickup_longitude']+train_data['dropoff_longitude'])/2))
train_data.head()
```
