## 유클리드 거리(벡터의 차의 길이)
### data science school pg 151
~~~
두 벡터가 가리키는 점 사이의 거리를 유클리드 거리라고 한다. 
두 벡터의 유클리드 거리는 벡터의 차의 길이로 구할 수 있다.

두 벡터의 내적은 두 벡터의 놈의 길이를 곱하고 두 벡터 사이의 각도의 코사인 함숫값으로 계산 가능하다.
a.T @ b = \\a\\ \\b\\ cos()
~~~

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([0, 1], [0, 2], 'k-', lw=3)     #'k-'는 검은색 선, lw는 line width의 약자로 선의 굵기 지정
plt.plot([0, 1], [0, 0], 'k-', lw=3)
plt.plot([1, 1], [0, 2], 'k-', lw=3)

plt.text(0.05, 1, 'h')
plt.text(0.35, -0.2, 'b')
plt.text(1.05, 1, 'a')
plt.text(0.12, 0.06, r'$\theta$')   #세타 그리는 방법(r 없으면 다른 결과 나옴)

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
         
plt.xlim(-1.1, 2.1)
plt.ylim(-0.5, 2.3)
plt.show()
```

## 코사인 그래프와 사인 그래프
### pg152

```python
x=np.linspace(0, np.pi/2, 100)        #linespace는 0~np.pi/2 범위의 균일한 간격의 숫자 100개 return
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x, y1, 'r--', lw=3, label=r'$\sin\theta$')
plt.plot(x, y2, 'b--', lw=3, label=r'$\cos\theta$')
plt.legend()                     #박스 안에 label을 그려주는 역할을 함

plt.xticks([0, np.pi/4, np.pi/2], [r'$0^{\circ}$', r'$45^{\circ}$', r'$90^{\circ}$'])
plt.xlabel(r'$\theta$')
plt.title(r'graph of $\sin\theta$ and $\cos\theta$')
```
