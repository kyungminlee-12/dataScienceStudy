```python
x=np.arange(-9, 10)
y=x**2

#linestyle은 line을 무엇으로 그릴지 알리는 keyword이고 marker는 y부분을 무슨 기호로 나타낼지 알려줌
plt.plot(x, y, linestyle=":", marker="*")   
plt.show()

#라인 스타일로는 '-', ':', '-.', '--' 등이 사용될 수 있습니다. 또한 X축 및 Y축에서 특정 범위를 자를 수도 있습니다.
```

```python
x = np.arange(-9, 10) 
y = x ** 2 

#plt.plot는 그래프를 그리는 함수이다. 
#이때 x, y를 먼저 입력하고 linestyle과 marker를 무엇으로 할 지 지정한다
plt.plot(x, y, linestyle="-.", marker="*")  
plt.xlim(-5, 5)                #x의 범위를 제한한다
plt.show()
```

```python
x = np.arange(-9, 10) 
y1 = x ** 2 
y2 = -x 

#plt.plot는 그래프를 생성하는 함수이다
#그래프의 x, y와 linestyle, marker, color, label 지정까지 가능하다
plt.plot(x, y1, linestyle="-.", marker="*", color="red", label="y = x * x") 
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x") 

#각 축의 단위를 써주는 것
plt.xlabel("X") 
plt.ylabel("Y") 
plt.legend( shadow=True, borderpad=1 )   #각 그래프가 무얼 나타내는 지 알려주는 표시

plt.show()    #최종적으로 시각화하여 보여주게 하는 코드
```

**plt.legend 사용법 (결론은 잘 모름)**
~~~
그래프가 무엇을 의미하하는지 박스 안에 설명해 놓은 것이 나타나게하는 코드

legend 함수는 레전드를 그래프에 위치하게 한다. 즉, 하나의 그래프 안에 위치하도록 설정한다.

레전드는 크게 3가지 방법으로 생성할 수 있다
1. 자동 기능을 이용한 작성
    legend()
    
2. 이미 이벽된 플롯 요소에 대한 라벨링을 통한 작성
    legend(labels)
    legend(labels=labels_data)    
    plt.plot(['A', 'B'])     #라벨명 직접 입력
    
    하지만 이 방법을 권장하지 않는다. 
    플롯 요소와 라벨간의 관계가 이 방법에 의해 불분명해지고, 순서가 섞일 가능성이 있다
    
3. 데이터 라벨의 직접 매칭
    legend(handles, labels)
~~~

**xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가 가능**
~~~
alpha는 투명도를 의미한다. 0이면 완전 투명, 1이면 완전 불투명이다
~~~

```python
import numpy as np
np.random.seed(0)

people = ['Bob', 'Taylor', 'Jacob', 'Min']
y_pos = np.arange(len(people))            #
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))       #len(people)은 random으로 뽑을 숫자 개수를 지정하는 것이다!!

plt.title("Barh Chart")

# plt.barh → 가로 막대 (기준들, 각 기준의 수치)
plt.barh(y_pos, performance, xerr=error, alpha=0.4)      #막대 그래프를 그리는 코드
plt.yticks(y_pos, people)                                #y축 내용 개수 지정 후 내용 지정
plt.xlabel('x level')
plt.show()
```

**2개 이상의 바 차트를 동시에 그리기**

```python
n_groups = 5
means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)
means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men',
                 align='edge')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women',
                 align='edge')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.tight_layout()
plt.show()
```

# 누적 막대그래프 생성

```python
x = np.random.rand(10)    # 아래 막대 
y = np.random.rand(10)    #중간 막대 
z = np.random.rand(10)    # 위 막대 

#누적 막대그래프에 담을 데이터 모두를 하나의 list 안에 넣기
data = [x, y, z]    
x_array = np.arange(10)    #(3개지 data의 개수들을 명시)

for i in range(0, 3):   # 누적 막대의 종류가 3개 
    plt.bar( x_array,    # 0부터 10까지의 X 위치에서  
            data[i],     # 각 높이(10개)만큼 쌓음 (각 데이터들을 list에 담긴 순서대로 쌓기)
            bottom=np.sum(data[:i], axis=0) )    #아래 있는 데이터들의 높이는 더한 값
    
plt.show()
```

**bottom 인수로 바의 위치를 조정하여 겹친 바 차트(stacked bar chart)도 그릴 수 있다.**

출처: https://namyoungkim.github.io/matplotlib/visualization/2017/10/07/visualization/


```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, color='r', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, color='y',
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

# Stem plot
~~~
바 차트와 유사하지만 폭(width)이 없는 스템 플롯(stem plot)도 있다.
주로 이산 확률 함수나 자기상관관계(auto-correlation)를 묘사할 때 사용된다.

documents는 아래서 확인.
~~~
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.stem


```python
x = np.linspace(0.1, 2 * np.pi, 10)

markerline, stemlines, baseline = plt.stem(x, np.cos(x), '-.')

plt.setp(markerline, 'markerfacecolor', 'b')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.show()
```

# pie chart
~~~
카테고리 별 값의 상대적인 비교를 해야 할 때는 pie으로 pie chart를 그려야 한다.
document
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie
~~~

```python
labels = u'rabbits', u'fishes', u'dogs', u'cats'

# 합은 100
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# 파이를 잘라서 떨어뜨리는 간격 조정
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

**s : scalar or array_like, shape (n, ), optional**
~~~
The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.
~~~

**c : color, sequence, or sequence of color, optional**
~~~
The marker color. Possible values:

    A single color format string.
    A sequence of color specifications of length n.
    A sequence of n numbers to be mapped to colors using cmap and norm.
    A 2-D array in which the rows are RGB or RGBA.
~~~

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
#c=colors 안 쓰면 다양한 색들이 나타나지 않음
#s=area 안 쓰면 동그라미 크기가 다양하지 않음 

plt.show()
```
