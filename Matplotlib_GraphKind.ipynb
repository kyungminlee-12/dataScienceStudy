```python
import numpy as np
```

```python
x=np.linspace(0, np.pi*10, 500)    #PI*10 넓이에, 500개의 점을 찍기
fig, axes=plt.subplots(2, 1)       #2개의 그래프가 들어가는 Figure 생성
axes[0].plot(x, np.sin(x))         #첫 번째 그래프는 사인 그래프
axes[1].plot(x, np.cos(x))         #두 번째 그래프는 코사인 그래프
fig.savefig("sin&cos.png")         #그림 파일으로 저장하기

#axes에 2개의 그래프를 index 형태로 표현할 수 있나봐 ( 내 추측)
# fig의 의미가 무엇일까?

```

# 그래프 그리기

**plt.figure()와 plt.show()를 위 아래에 쓴다는 가정 하에**

 ~~~
 plt.plot([1, 2, 3], [1, 2, 3], marker='*')     #marker는 특정 좌표에 찍히는 점
 plt.scatter([1,2,3], [1,2,3])                  #선이 연결되지 않고 점만 표시
 plt.bar([1,2,3], [1,2,3])                      #막대 그래프
 plt.pie([1,2,3])                               #원형 그래프
 ~~~

```python
x=np.arange(-9, 10)
y=x**2

#linestyle은 line을 무엇으로 그릴지 알리는 keyword이고 marker는 y부분을 무슨 기호로 나타낼지 알려줌
plt.plot(x, y, linestyle=":", marker="*")   
plt.show()

#라인 스타일로는 '-', ':', '-.', '--' 등이 사용될 수 있습니다. 또한 X축 및 Y축에서 특정 범위를 자를 수도 있습니다.
```

```python
x = np.arange(-9, 10) 
y = x ** 2 

#plt.plot는 그래프를 그리는 함수이다. 
#이때 x, y를 먼저 입력하고 linestyle과 marker를 무엇으로 할 지 지정한다
plt.plot(x, y, linestyle="-.", marker="*")  
plt.xlim(-5, 5)                #x의 범위를 제한한다
plt.show()
```

```python
x = np.arange(-9, 10) 
y1 = x ** 2 
y2 = -x 

#plt.plot는 그래프를 생성하는 함수이다
#그래프의 x, y와 linestyle, marker, color, label 지정까지 가능하다
plt.plot(x, y1, linestyle="-.", marker="*", color="red", label="y = x * x") 
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x") 

#각 축의 단위를 써주는 것
plt.xlabel("X") 
plt.ylabel("Y") 
plt.legend( shadow=True, borderpad=1 ) 

plt.show()    #최종적으로 시각화하여 보여주게 하는 코드
```

**plt.legend 사용법 (결론은 잘 모름)**
~~~
legend 함수는 레전드를 그래프에 위치하게 한다. 즉, 하나의 그래프 안에 위치하도록 설정한다.

레전드는 크게 3가지 방법으로 생성할 수 있다
1. 자동 기능을 이용한 작성
    legend()
    
2. 이미 이벽된 플롯 요소에 대한 라벨링을 통한 작성
    legend(labels)
    legend(labels=labels_data)    
    plt.plot(['A', 'B'])     #라벨명 직접 입력
    
    하지만 이 방법을 권장하지 않는다. 
    플롯 요소와 라벨간의 관계가 이 방법에 의해 불분명해지고, 순서가 섞일 가능성이 있다
    
3. 데이터 라벨의 직접 매칭
    legend(handles, labels)
~~~

# 막대 그래프 생성

```python
x = np.arange(-9, 10) 

#plt.bar(x_arange, y_arange)
plt.bar(x, x ** 2)    #처음은 x arange, 두 번째는 y arange를 입력한다.
plt.show()
```

# 누적 막대그래프 생성

```python
x = np.random.rand(10)  # 아래 막대 
y = np.random.rand(10)  #중간 막대 
z = np.random.rand(10)  # 위 막대 

#누적 막대그래프에 담을 데이터 모두를 하나의 list 안에 넣기
data = [x, y, z]    
x_array = np.arange(10)    #(3개지 data의 개수들을 명시)

for i in range(0, 3):   # 누적 막대의 종류가 3개 
    plt.bar( x_array,    # 0부터 10까지의 X 위치에서  
            data[i],     # 각 높이(10개)만큼 쌓음 (각 데이터들을 list에 담긴 순서대로 쌓기)
            bottom=np.sum(data[:i], axis=0) )    #아래 있는 데이터들의 높이는 더한 값
    
plt.show()
```

# 스캐터(scatter) 그래프
~~~
그래프를 그릴 때 3번째 인자로 마커의 종류를 입력하면, 알아서 스캐터 그래프로 인식합니다.
~~~

```python
x = np.arange(-9, 10) 
y1 = x ** 2 
y2 = -x 

plt.plot( x, y1, "*", markersize=10, markerfacecolor="blue", markeredgecolor="red" ) 
plt.show()
```
